```python
df["Order.Date"] = pd.to_datetime(df["Order.Date"])
df["Ship.Date"] = pd.to_datetime(df["Ship.Date"]) # Convert data columns
```

```python
df["Quantity"] = df["Quantity"].astype(int) # Convert Integer Columns
```

```python
df["Sales"] = df["Sales"].astype(float)
df["Profit"] = df["Profit"].astype(float) # Convert Decimal Columns
```

```python
df["Category"] = df["Category"].astype("category")
df["Segment"] = df["Segment"].astype("category")
df["Region"] = df["Region"].astype("category")
df["Ship.Mode"] = df["Ship.Mode"].astype("category") # Convert Categorical Columns
```

```python
df.to_csv("cleaned_superstore.csv", index=False) # Save the Cleaned Dataset
```
